```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, DBSCAN
from sklearn.cluster import Birch
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Define clustering algorithms
kmeans = KMeans(n_clusters=3)
dbscan = DBSCAN(eps=0.5, min_samples=5)
birch = Birch(n_clusters=3)
agg_clustering = AgglomerativeClustering(n_clusters=3)
gmm = GaussianMixture(n_components=3)

# Fit and predict clusters
kmeans_labels = kmeans.fit_predict(X)
dbscan_labels = dbscan.fit_predict(X)
birch_labels = birch.fit_predict(X)
agg_clustering_labels = agg_clustering.fit_predict(X)
gmm_labels = gmm.fit_predict(X)

# Plotting the results
plt.figure(figsize=(12, 12))

plt.subplot(3, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=300, c='red')
plt.title('K-means')

plt.subplot(3, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN')

plt.subplot(3, 2, 3)
plt.scatter(X[:, 0], X[:, 1], c=birch_labels, cmap='viridis')
plt.title('Birch')

plt.subplot(3, 2, 4)
plt.scatter(X[:, 0], X[:, 1], c=agg_clustering_labels, cmap='viridis')
plt.title('Agglomerative Clustering')

plt.subplot(3, 2, 5)
plt.scatter(X[:, 0], X[:, 1], c=gmm_labels, cmap='viridis')
plt.title('GMM')

plt.show()
```
